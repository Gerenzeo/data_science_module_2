```python
import pandas as pd
```

```python
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
table_12 = pd.read_html(url)[12]
```

1. Вивести перші рядки таблиці за допомогою методу head

```python
print(table_12.head())
```

2. Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
print(table_12.shape)
```

3. Замініть у таблиці значення "—" на значення NaN

```python
table_12.replace("—", "NaN", inplace=True)
print(table_12)
```

4. Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
print(table_12.dtypes)
```

5. Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
table_12.replace("NaN", 0, inplace=True)
print(table_12)
```

6. Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
print(table_12.isnull().sum() / len(table_12) * 100)
```

7. Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
table_12.drop(table_12.index[-1], inplace=True)
print(table_12)
```

8. Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
table_12.fillna(table_12.mean(numeric_only=True), inplace=True)
print(table_12)
```

9. Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
table_12["2019"] = pd.to_numeric(table_12["2019"], errors="coerce")
average_2019 = table_12["2019"].mean()

print(table_12[table_12['2019'] > average_2019])
```

10. У якому регіоні була найвища народжуваність у 2014 році?

```python
table_12["2014"] = pd.to_numeric(table_12["2014"], errors="coerce")
print(table_12.loc[table_12["2014"].idxmax(), "Регіон"])
```

11. Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
import matplotlib.pyplot as plt
```

```python
data_2019 = table_12[["Регіон", "2019"]]
data_2019 = data_2019.dropna()
data_2019["2019"] = pd.to_numeric(data_2019["2019"], errors="coerce")

sort_data = data_2019.sort_values('2019', ascending=False)

plt.figure(figsize=(12, 8))
plt.bar(sort_data["Регіон"], sort_data['2019'], color='navy')
plt.title("Народжуваність по регіонах в 2019 році")
plt.xlabel("Регіон")
plt.ylabel("Кількість народжень в 2019 році")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
